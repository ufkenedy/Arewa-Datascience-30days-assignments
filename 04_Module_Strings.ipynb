```python

### Exercise 1: Concatenate strings


concatenated_string = ' '.join(['Thirty', 'Days', 'Of', 'Python'])
print(concatenated_string)


### Exercise 2: Concatenate another set of strings


concatenated_string_2 = ' '.join(['Coding', 'For', 'All'])
print(concatenated_string_2)


### Exercise 3: Declare and print a variable


company = "Coding For All"
print(company)


### Exercise 4: Print the length of the string


print(len(company))


### Exercise 5: Change all characters to uppercase


print(company.upper())


### Exercise 6: Change all characters to lowercase


print(company.lower())


### Exercise 7: Use different string methods


print(company.capitalize())
print(company.title())
print(company.swapcase())


### Exercise 8: Slice out the first word


first_word_removed = company[company.index(" ")+1:]
print(first_word_removed)


### Exercise 9: Check if the string contains a word


print('Coding' in company)
print(company.index('Coding'))
print(company.find('Coding'))


### Exercise 10: Replace a word in the string


new_company = company.replace('Coding', 'Python')
print(new_company)


### Exercise 11: Change a phrase using replace


phrase = "Python for Everyone"
new_phrase = phrase.replace("Everyone", "All")
print(new_phrase)


### Exercise 12: Split the string


split_company = company.split()
print(split_company)

### Exercise 13: Split a string at commas

tech_companies = "Facebook, Google, Microsoft, Apple, IBM, Oracle, Amazon"
split_tech = tech_companies.split(', ')
print(split_tech)


### Exercise 14: Character at index 0


print(company[0])

### Exercise 15: Last index of the string


print(len(company) - 1)


### Exercise 16: Character at index 10


print(company[10])


### Exercise 17: Create an acronym


phrase1 = "Python For Everyone"
acronym1 = ''.join([word[0] for word in phrase1.split()])
print(acronym1)

phrase2 = "Coding For All"
acronym2 = ''.join([word[0] for word in phrase2.split()])
print(acronym2)


### Exercise 18: Index of first occurrence of 'C'


print(company.index('C'))


### Exercise 19: Index of first occurrence of 'F'


print(company.index('F'))


### Exercise 20: Last occurrence of 'l'


sentence = "Coding For All People"
print(sentence.rfind('l'))


### Exercise 21: First occurrence of 'because'


sentence = 'You cannot end a sentence with because because because is a conjunction'
print(sentence.find('because'))


### Exercise 22: Last occurrence of 'because'


print(sentence.rindex('because'))


### Exercise 23: Slice out 'because because because'


first_because = sentence.find('because')
last_because = sentence.rfind('because') + len('because')
sliced_sentence = sentence[:first_because] + sentence[last_because:]
print(sliced_sentence)


### Exercise 24: Check if string starts with a substring


print(company.startswith('Coding'))


### Exercise 25: Check if string ends with a substring


print(company.endswith('coding'))


### Exercise 26: Remove trailing spaces

trimmed_company = '   Coding For All      '.strip()
print(trimmed_company)


### Exercise 27: Check if variable is an identifier

print('30DaysOfPython'.isidentifier())
print('thirty_days_of_python'.isidentifier())


### Exercise 28: Join list with hash

libraries = ['Django', 'Flask', 'Bottle', 'Pyramid', 'Falcon']
joined_libraries = ' # '.join(libraries)
print(joined_libraries)


### Exercise 29: New line escape sequence


print("I am enjoying this challenge.\nI just wonder what is next.")


### Exercise 30: Tab escape sequence


print("Name\tAge\tCountry\tCity")
print("Asabeneh\t250\tFinland\tHelsinki")


### Exercise 31: String formatting for area


radius = 10
area = 3.14 * radius ** 2
print(f"The area of a circle with radius {radius} is {area} meters square.")

### Exercise 32: String formatting for arithmetic operations

a, b = 8, 6
print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b:.2f}")
print(f"{a} % {b} = {a % b}")
print(f"{a} // {b} = {a // b}")
print(f"{a} ** {b} = {a ** b}")

### Final Message
```
